```python
import cv2
import os
import numpy as np
import pandas as pd

# Tentukan path folder tempat citra disimpan
folder_path = '/content/drive/MyDrive/Dataset Jeruk/Test'

# Buat list kosong untuk menyimpan nilai fitur warna dari setiap citra
features = []
# Buat list kosong untuk menyimpan nama file citra
image_names = []

# Loop untuk membaca setiap citra di dalam folder
for filename in os.listdir(folder_path):
    # Baca citra
    img = cv2.imread(os.path.join(folder_path, filename))
    # Hitung histogram untuk setiap kanal warna
    hist_r = cv2.calcHist([img], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([img], [1], None, [256], [0, 256])
    hist_b = cv2.calcHist([img], [2], None, [256], [0, 256])
    # Gabungkan histogram untuk setiap kanal warna menjadi satu array fitur warna
    color_feature = np.concatenate((hist_r, hist_g, hist_b), axis=None)
    # Tambahkan fitur warna ke dalam list
    features.append(color_feature)
    # Tambahkan nama file citra ke dalam list
    image_names.append(filename)

# Konversi list fitur warna dan nama file citra menjadi numpy array
features = np.array(features)
image_names = np.array(image_names)

# Buat DataFrame dari numpy array fitur warna dan nama file citra
df = pd.DataFrame({'Image': image_names, 'R': features[:,0], 'G': features[:,1], 'B': features[:,2]})

# Tampilkan DataFrame di layar
print(df)
```

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB

# Membaca data
data = pd.read_csv("/content/drive/MyDrive/Dataset Jeruk/Hasil/fitur_warna.csv")

# Ambil fitur RGB
X = data[["R", "G", "B"]]

# Tentukan 2 threshold
threshold1 = 4500
threshold2 = 10000

# Membuat label 3 kelas
y = []
for rgb in X.values:
    total = sum(rgb)
    if total < threshold1:
        y.append("Kuning")
    elif total < threshold2:
        y.append("Oranye")
    else:
        y.append("Hijau")

# Model Naive Bayes
clf = GaussianNB()
clf.fit(X, y)

# Prediksi
pred = clf.predict(X)

# Simpan hasil
data["Warna"] = pred

# Tampilkan
print(data)
```

```python
import cv2
import numpy as np
import pandas as pd
from skimage.feature import graycomatrix, graycoprops
import os
```

```python
# list properti yang akan diekstraksi
maturity_properties = ['hue_mean', 'saturation_mean', 'value_mean', 'hue_std', 'saturation_std', 'value_std', 'texture_contrast']

# path folder yang berisi gambar
folder_path = '/content/drive/MyDrive/Dataset Jeruk/Test/'
data = []
```

```python
# loop untuk setiap file citra di folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.jpg') or file_name.endswith('.jpeg') or file_name.endswith('.png') or file_name.endswith('.bmp'):
        row_data = []
        file_path = os.path.join(folder_path, file_name)
        row_data.append(file_name)

        # preprocessing gambar
        src = cv2.imread(file_path)
        src = cv2.resize(src, (300, 300))
        hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)
        h, s, v = cv2.split(hsv)

        # ekstraksi properti warna
        hue_mean = np.mean(h)
        saturation_mean = np.mean(s)
        value_mean = np.mean(v)
        hue_std = np.std(h)
        saturation_std = np.std(s)
        value_std = np.std(v)

        # ekstraksi properti bentuk
        contours, hierarchy = cv2.findContours(cv2.Canny(src, 100, 200), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        cnt = max(contours, key=cv2.contourArea)
        perimeter = cv2.arcLength(cnt, True)
        area = cv2.contourArea(cnt)
        circularity = 4 * np.pi * (area / perimeter ** 2)
        aspect_ratio = float(src.shape[0]) / src.shape[1]
        extent = area / (src.shape[0] * src.shape[1])
        solidity = area / cv2.contourArea(cv2.convexHull(cnt))

        # hitung kontras tekstur
        g = graycomatrix(v, distances=[1], angles=[0], symmetric=True, normed=True)
        contrast = graycoprops(g, 'contrast')
        # tambahkan ke data
        row_data.extend([hue_mean, saturation_mean, value_mean, hue_std, saturation_std, value_std, contrast])
        data.append(row_data)
```

```python
# export ke file Excel
df = pd.DataFrame(data, columns=['file_name'] + maturity_properties)
file_path = os.path.join(folder_path, '2218083EkstratsiFiturPhyton.xlsx')
df.to_excel(file_path, index=False)
```

```python
def klasifikasi_warna(hue):
    if hue < 35:
        return "Oranye (Matang)"
    elif hue < 70:
        return "Hijau (Mentah)"
    else:
        return "Kuning (Setengah Matang)"
```
